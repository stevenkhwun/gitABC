* The Raw URL of (e.g. https://raw.githubusercontent.com) the GitHub URL should be used instead of the standard GitHub URL (e.g. https://github.com) of the image. You can get the Raw URL by right-click on the GitHub page displaying your image and select "Open image in new tab", and then copy the URL from the new tab's address bar.

* Alternatively, you can use the syntax `![Alt text](url)`, which is a Markdown format used to embed images into a document or web page. The Alternative Text (Alt text) inside the square bracket is a brief description of the image for accessibility.
* This method may not able to control the size of the image.
* The following code embed the image to the notebook

```markdown
![Alt text](https://raw.githubusercontent.com/stevenkhwun/gitABC/refs/heads/main/IMG_0205_A.png)
*Moji and Bailey*
```
* The `*Moji and Bailey*` is added as an alternative way to add a caption to the image.

![Alt text](https://raw.githubusercontent.com/stevenkhwun/gitABC/refs/heads/main/IMG_0205_A.png)

* Finally, you can also put the image and text in a single-column table to keep them grouped.
* The code used is as follow:

```markdown

| :---: |
| Moji and Bailey |
```
